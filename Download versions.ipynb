```python
from bs4 import BeautifulSoup
import requests
import os

url = 'http://archive.apache.org/dist/cxf/'
ext = ''

def listFD(url, ext=''):
    page = requests.get(url).text
    #print(page)
    soup = BeautifulSoup(page, 'html.parser')
    return [url + '/' + node.get('href') for node in soup.find_all('a') if node.get('href').endswith(ext)]

for directory in listFD(url, ext):
    print("wget %s" %directory)
    for file in listFD(directory, "zip"):
        print("wget %s" %file)
        if (file.endswith("src.zip")):
            continue
        elif (file.endswith("zip")):
            os.system("wget %s" %file)
```
